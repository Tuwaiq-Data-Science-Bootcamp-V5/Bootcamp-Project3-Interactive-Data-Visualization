# Data Cleaning

### Column Names and Types

```python
# drop unwanted column

df.drop(df.filter(regex="Unnamed"), axis=1, inplace=True)
```

```python
# rename for clarity

df.rename(columns={'neighbourhood':'district','space':'area'},inplace=True)
```

```python
# change types

changetype = ["stairs","tent", "driverRoom","patio", "kitchen", "outdoorRoom", "maidRoom", "garage","duplex", "elevator", "furnihsed","pool", "basement"]

for col in changetype:
    df[col] = df[col].replace([1, 0], [True, False])
    df[col] = df[col].astype(bool)
```

```python
# change Arabic values to English (front)
print('Before: \n',df['front'].value_counts())
df['front'] = df['front'].replace(['شمالية', 'شرقية', 'غربية', 'جنوبية','شمالية شرقية','جنوبية غربية','جنوبية شرقية', 'شمالية غربية', 'ثلاث شوارع', 'أربع شوارع']\
                                  , ['Northern', 'Eastern', 'Western', 'Southern', 'Northeastern', 'Southwestern', 'Southeastern', 'Northweatern', 'Three Streets', 'Four Streets'])
print('-----------\nAfter: \n',df['front'].value_counts())
```

```python
# change Arabic values to English (location)
print('Before: \n',df['location'].value_counts())
df['location'] = df['location'].replace(['غرب الرياض', 'شرق الرياض', 'شمال الرياض', 'جنوب الرياض','وسط الرياض']\
                                  , ['West', 'East', 'North', 'South', 'Central'])
print('-----------\nAfter: \n',df['location'].value_counts())
```

```python
# lounges

print('Before: \n',df['lounges'].value_counts())

# Change unwanted value detail for ease
df['lounges'] = df['lounges'].replace('7+', 7)

# Convert column to numeric, handling errors by coercing NaN
df['lounges'] = pd.to_numeric(df['lounges'], errors='coerce')

# Convert the column to integers, rounding float values
df['lounges'] = df['lounges'].astype('Int64')

# Display the updated value counts
print('-----------\nAfter: \n',df['lounges'].value_counts())
```

```python
# apartments

print('Before: \n',df['apartments'].value_counts())

df['apartments'] = df['apartments'].replace('30+', 30)

df['apartments'] = pd.to_numeric(df['apartments'], errors='coerce')

df['apartments'] = df['apartments'].astype('Int64')

print('-----------\nAfter: \n',df['apartments'].value_counts())
```

```python
# bathrooms

print('Before: \n',df['bathrooms'].value_counts())

df['bathrooms'] = df['bathrooms'].replace('5+', 5)

df['bathrooms'] = pd.to_numeric(df['bathrooms'], errors='coerce')

df['bathrooms'] = df['bathrooms'].astype('Int64')

print('-----------\nAfter: \n',df['bathrooms'].value_counts())
```

### Handelling Missing Data 

```python
# streetWidth 

print('Before: \n',df['streetWidth'].describe().T)

# Imputation by average streetWidth to area ration
stwidth_to_area_ratio = df['streetWidth'] / df['area']

average_stwidth_to_area_ratio = stwidth_to_area_ratio.mean()

df['streetWidth'] = df['streetWidth'].fillna(df['area']*average_stwidth_to_area_ratio)

print('-----------\nAfter: \n',df['streetWidth'].describe().T)
```

- Imputed them by the median lounges per no. of rooms

```python
# lounges

print('Before: \n',df['lounges'].describe().T)

# Imputation by median of lounges to no. rooms
median_lounges_rooms = df.groupby('rooms')['lounges'].median()

df['lounges'] = df['lounges'].fillna(df['rooms'].map(median_lounges_rooms))

df['lounges'] = df['lounges'].astype('Int64')

print('-----------\nAfter: \n',df['rooms'].describe().T)
```

```python
# price and square price

print('Before: \n',df['price'].describe().T)

# Imputation by average square price in district
mean_sq_price_district = df.groupby('district')['square price'].mean()

df['square price'].fillna(df['district'].map(mean_sq_price_district), inplace=True)

# Imputation by square price * area
df['price'].fillna(df['area']*df['square price'], inplace=True)


print('-----------\nAfter: \n',df['price'].describe().T)
```
