## Import library

```python
import numpy as nd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
```

## Read and Show Data

```python
RYD = pd.read_csv('RiyadhVillasAqar.csv')
```

```python
RYD
```

## EDA

```python
RYD.info()
```

```python
RYD['lounges']=RYD['lounges'].str.replace('+', '').astype(float)
RYD['bathrooms']=RYD['bathrooms'].str.replace('+', '').astype(float)
RYD['apartments']=RYD['apartments'].str.replace('+', '').astype(float) # <----- This code cause NaN Value to the apartments columns 
```

```python
RYD['apartments'].fillna(0, inplace=True) # <------ I check the values with loop and if statement and there is no null value before changing dtype to float but there is after change it so 
#I fill it with 0 insted of droping the value
```

```python
RYD.duplicated().value_counts()
```

```python
RYD.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
RYD.isna().sum()
```

```python
RYD['lounges'].fillna(RYD['lounges'].median(), inplace= True) # <------- write a better way if you have
RYD['streetWidth'].fillna(RYD['streetWidth'].median(), inplace= True) # <------- write a better way if you have 
```

```python
RYD.describe().T
```

```python
RYD.dropna(inplace=True)
```

```python
RYD.isna().sum()
```

## Data Visualization

```python
x = RYD.select_dtypes(include=['number'])
plt.figure(figsize=(25, 20))
sns.heatmap(x.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Delete it if it's not useful
x = ['price', 'space' ,'rooms', 'bathrooms', 'lounges' , 'pool']
sns.pairplot(RYD[x])
plt.show()
```

```python
fig = px.scatter(RYD, x='neighbourhood', y='price')
fig.update_layout(title_text='Price by Neighbourhood', title_x=0.5)
fig.update_xaxes(tickangle=90)
fig.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(RYD ,x='duplex', y='price', hue='location')
plt.show()
```

```python
plt.figure(figsize=(6,3))
x = RYD.groupby(['location','pool']).mean('price').reset_index()
fig = px.bar(x, x='location',y='price')
fig.update_layout(title_text='Price Average with pool', title_x=0.5)
fig.show()
```

```python
x = RYD.groupby('propertyAge')['price'].mean().reset_index()
fig = px.line(x, x='propertyAge', y='price')
fig.update_layout(title_text='Real Estate Age and Average Price', title_x=0.5)
fig.show()
```

```python
px.histogram(RYD,x=RYD['neighbourhood'],y=RYD['price'])
fig.update_layout(title_text='Neighbourhood Price ', title_x=0.5)
fig.show()
```
